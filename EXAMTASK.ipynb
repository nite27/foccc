In a new Code block, write a function that takes as its parameter a list of strings. The function should return
True if one of the strings is coconut (in lower case, as shown), or False otherwise. Include some code
outside the function that shows that the function works correctly. Again, you may assume the "Happy Path",
so the list will always contain just strings.

```python
def has_coconut(strings):
    return "coconut" in strings

# Test cases
print(has_coconut(["apple", "banana", "coconut"])) 
print(has_coconut(["apple", "banana", "orange"]))   
print(has_coconut(["coconut"]))                      
print(has_coconut(["Coconut", "apple"]))           
```

TASK 5:

In a last new Code block, write a program that prompts the user to enter strings, one at a time. The input
ends when the user just presses "Enter" (so the string entered is empty). Then the program should display
the strings, in lower case, in alphabetical order.


```python
strings = []

while True:
    user_input = input("Enter a string (or press Enter to finish): ")
    if user_input == "":
        break
    strings.append(user_input)



sorted_strings = sorted([s.lower() for s in strings])

print("\nStrings in lowercase, alphabetical order:")
for s in sorted_strings:
    print(s)
```
